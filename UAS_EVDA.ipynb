```python
import matplotlib.pyplot as plt
tahunmasuk=data1str['Tahun Masuk'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(tahunmasuk.index, tahunmasuk.values, color='skyblue')
plt.xlabel('Tahun masuk')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Tahun Masuk')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
tahunwisuda=data1str['Tahun Wisuda'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(tahunwisuda.index, tahunwisuda.values, color='skyblue')
plt.xlabel('Tahun Wisuda')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Tahun Wisuda')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
waktulamadapatkerja=data1str['Waktu Mendapatkan Pekerjaan:'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(waktulamadapatkerja.index, waktulamadapatkerja.values, color='skyblue')
plt.xlabel('Waktu Mendapatkan Pekerjaan')
plt.ylabel('Jumlah Mahasiswa')
plt.title('Waktu Mahasiswa Mendapatkan Pekerjaan')
plt.xticks(rotation=45)
plt.show()
```
